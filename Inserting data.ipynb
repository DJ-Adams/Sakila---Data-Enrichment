df.to_sql()
Just as Pandas has both a pd.read_csv() function and a df.to_csv() method, it also has a pd.read_sql() function (which you have been using) AND a df.to_sql() method.

We will use the .to_sql method to add our data, but we must make sure to select the correct arguments for the task.

If we check the documentation for to_sql, we can see that:
name: it needs the name of the table to insert the data into
con: it needs the connection to the database from sqlalchemy
schema: It does not require that we specify a schema for MySQL.
if_exists: By default, if our table already exists, it will fail, and we will get an error.

This is not what we want! We want to add our data TO the table if it already exists.

The options for if_exists are:
fail: Raise a ValueError.
replace: Drop the table before inserting new values.
append: Insert new values to the existing table.
Since we want to add our data to the database, will use if_exists='append'
index: it includes the DataFrame index as a column for the database.
We will generally want to have a plain pandas integer index for our DataFrame and use index=False.
We will revisit some of the remaining arguments later as they become necessary.

```python
employees.to_sql("employee",conn,index=False, if_exists='append')
```

```python
# Checking the setting for FOREIGN_KEY_CHECKS
q = """SELECT @@FOREIGN_KEY_CHECKS"""
pd.read_sql(q, conn)
```

```python
# Changing the setting for FOREIGN_KEY_CHECKS with pd.read_sql
q = """SET @@FOREIGN_KEY_CHECKS=0"""
pd.read_sql(q,conn) 
```

Sqlalchemy automatically closed our connection due to this error. To get around this, we can use the connection itself to conn.execute() the query.

Whenever we need to run a query that does not return data, we will need to use the connection object to .execute the query instead of using pd.read.sql.

```python
# Changing the setting for FOREIGN_KEY_CHECKS with the connection
q = """SET @@FOREIGN_KEY_CHECKS=0"""
conn.execute(q)
```

```python
# Confirm the checks are deactiavated
q = """SELECT @@FOREIGN_KEY_CHECKS"""
pd.read_sql(q,conn)
```

```python
# Inserting the data now that foreign key checks are disabled
employees.to_sql("employee",conn,index=False, if_exists='append')
```

Duplicate Key Errors
Let's say you received a new set of data to add to the database. First, let's load in the new file and perform the same preprocessing steps.

```python
# Loading new file and repeating preprocessing/cleaning steps
employees = pd.read_csv("Employee-data-v2.csv")
# replace original column names
employees.columns = employees.columns.str.upper()
# Rename EMPLOYEE_ID to "EMP_ID"
rename_map = {"EMPLOYEE_ID":"EMP_ID",
             "BIRTH_DATE":"B_DATE"}
employees = employees.rename(rename_map, axis=1)
# Changing B_DATE to datetime dtype
employees['B_DATE'] = pd.to_datetime(employees['B_DATE'])
employees
```

Since we closed and reopened the connection, we must again disable the foreign keys.

```python
# Disabling foreign key checks again
q = """SET @@FOREIGN_KEY_CHECKS=0"""
conn.execute(q)
```

We will now run into a new issue because adding version 2 of the csv file would result in duplicate primary keys, so we will get an IntegrityError about duplicate entries.

```python
# Attempt the insertion again after deactivating the checks
employees.to_sql("employee",conn,index=False, if_exists='append')
```

Unfortunately, if any of the primary keys being added are duplicate entries, the entire transaction fails, and no rows are added to the table.

```python
# Confirming no new rows have been added
q = """SELECT * FROM employee"""
pd.read_sql(q, conn)
```

Modifying df.to_sql()
Thus far, we have always used the if_exists='append' argument. We can change to using if_exists='replace', but it's important to realize that this will replace the entire table. It will not just replace rows that exist.

Critically, we cannot add if_exists='replace' alone. Doing so would cause us to completely drop the existing table, including its specified data types.

The Dangers of if_exists='replace'
Let's check the Types of the Employee table. Then, we will use a to_sql command with if_exists='replace', and then we will check the types again to see if they're the same.

```python
# Checking the data types before inserting new data
q = """DESCRIBE employee;"""
pd.read_sql(q,conn)
```

Now let's see what happens if we upload use if_exists='replace'. DO NOT run the code below! It is only for demonstration purposes.
employees.to_sql("EMPLOYEE", conn, index=False, if=exists='replace')

q= """DESCRIBE EMPLOYEE;"""
pd.read_sql(q,conn)

1) Creating a data type dictionary
SQLAlchemy has a solution to this issue. There is a types module in sqlalchemy that contains classes for each of the SQL data types. We can create a dictionary with each column as a key and the corresponding SQL data type class as the values.

We will then use this dictionary as the dtype argument for df.to_sql().

Let's walk through an example.

If we review our original DESCRIBE EMPLOYEE, we had the following data types: - varchar(45) - char(9) - decimal(10,1) - date - datetime Each of these has a corresponding class in the "sqlalchemy.types" module.

```python
from sqlalchemy.types import VARCHAR, CHAR, DECIMAL, DATE, DATETIME
```

employee_dtypes  = {'EMP_ID':VARCHAR(9),
                    'F_NAME':VARCHAR(45),
                    'L_NAME':VARCHAR(45),
                    'SSN': CHAR(9),
                    'B_DATE':DATE,
                    'SEX':CHAR(1),
                    'ADDRESS':VARCHAR(45),
                    'JOB_ID':CHAR(9),
                    'SALARY':DECIMAL(10,2),
                    'MANAGER_ID':CHAR(9),
                    'DEP_ID':CHAR(9),
                    'created_at':DATETIME,
                    'updated_at':DATETIME}
employee_dtypes

2) Run df.to_sql with the types dict as the dtype argument
Use this dictionary as the dtype argument in df.to_sql (make sure index=False)

```python
# Changing if_exists to replace to demo consequences.
employees.to_sql("employee",conn, index=False, if_exists='replace',
                dtype=employee_dtypes)
# Checking the describe again
q = """DESCRIBE employee;"""
pd.read_sql(q,conn)
```

We have the same data types after using if_exists='replace' and providing a dtype dictionary.

Notice that we no longer have "created_at" and "updated_at" columns. Even though we included them in our dictionary, we did not have any columns in our data frame, so they were not created.

Adding Back Created At and Updated At
